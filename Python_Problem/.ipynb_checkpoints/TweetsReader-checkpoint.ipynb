```python
import tweepy
import json
```

# Credentials

```python
consumer_key = "" 
consumer_secret = ""
access_key = ""
access_secret = ""
```

```python
# Extract tweets in JSON Format
def get_tweets_in_json_frmt(twitter_handle): 
          
        # Authorization to consumer key and consumer secret 
        auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
  
        # Access to user's access key and access secret 
        auth.set_access_token(access_key, access_secret) 
  
        # Calling api 
        api = tweepy.API(auth) 
  
        # Extract maximum 500 tweets
        number_of_tweets=500
        tweets = api.user_timeline(screen_name=twitter_handle) 
          
        json_res = {}
        for tweet in tweets:
            json_res[tweet.id_str] = {}
            json_res[tweet.id_str]['text'] = tweet.text
            json_res[tweet.id_str]['date'] = tweet.created_at.strftime('%d/%m/%Y')
            json_res[tweet.id_str]['likes'] = tweet.favorite_count
            json_res[tweet.id_str]['retweets'] = tweet.retweet_count
            
            count = 0
            if 'media' in tweet.entities:
                for medium in tweet.entities['media']:
                    if medium['type'] == 'photo':
                        count += 1
                        
            json_res[tweet.id_str]['image_count'] = count
  
        # retuen tweets
        return json_res
```

```python
json_res = get_tweets_in_json_frmt("@midasIIITD")
with open('output.json', 'w') as outfile:  
    json.dump(json_res, outfile)
```
